```python
import numpy as np
from matplotlib import pyplot as plt
from math import *
```

```python
## Constants
kb = 8.617333*10**-5 #eV/K
kb_J = 1.380649*10**-23
hbar = 1.054571*10**-34  #J*s
he = 6.582119*10**-16 #eV*s
h_J = 6.626 * 10**-34 #J*s
c = 2.99792*10**8 #m/s
```

```python
# 1a

temps = np.linspace(10,50000, 500)

acoustic = 1/(np.exp(.16/(kb*temps) -1))
optical = 1/(np.exp(.21/(kb*temps) -1))

print(f'300K acoustic = {1/(np.exp(.16/(kb*300) ) -1)}')
print(f'3000K acoustic = {1/(np.exp(.16/(kb*3000) ) -1)}')
print(f'300K optical = {1/(np.exp(.21/(kb*300) ) -1)}')
print(f'300K optical = {1/(np.exp(.21/(kb*3000) ) -1)}')

1/(np.exp(.21/(kb*temps) )-1)

plt.figure(1)
plt.plot(temps, acoustic)
plt.plot(temps, optical)
plt.legend(['Acoustic','Optical'])
plt.xlabel('Temp (K)')
plt.ylabel('f BE')
plt.show()
```

```python
#1b
E = [1, 11.5, 20] #eV
Ef = 11.7  #eV
E_sweep = np.linspace(8, 20, 200)
temps = np.linspace(1, 3000, 8)
temp_labels = [str(round(temp,2)) + ' K' for temp in temps]

# T = 1
print(f'1K E=1 = {1/(np.exp( (1-Ef) / ((kb*1))) +1)}')
print(f'1K E=11.5 = {1/(np.exp( (11.5-Ef) / ((kb*1))) +1)}')
print(f'1K E=20 = {1/(np.exp( (20-Ef) / ((kb*1)))  +1)}')

# T = 3000K
print(f'3000K E=1 = {1/(np.exp( (1-Ef) / ((kb*3000))) +1)}')
print(f'3000K E=11.5 = {1/(np.exp( (11.5-Ef) / ((kb*3000))) +1)}')
print(f'3000K E=20 = {1/(np.exp( (20-Ef) / ((kb*3000))) +1)}')



values = {}
for temp in temps:
    
    values[temp] = ( 1/(np.exp( (E_sweep-Ef) / ((kb*temp))) +1) )

plt.figure()
for temp in temps:
    plt.plot(E_sweep, values[temp])
plt.legend(temp_labels)
plt.xlabel('Energy (eV)')
plt.ylabel('f FD')
plt.annotate('Fermi Energy = 11.5eV', (3, .01))
plt.show()
```
